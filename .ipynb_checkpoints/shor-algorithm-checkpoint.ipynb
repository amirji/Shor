The shore's algorithm itself:

1. Pick a pseudo-random number a < N
2. Compute gcd(a, N). This may be done using the Euclidean algorithm.
3. If gcd(a, N) ≠ 1, then there is a nontrivial factor of N, so we are done.
4. Otherwise, use the period-finding subroutine (below) to find r, the period of the following function:
5. f(x) = ax mod N, i.e. the smallest integer r for which f(x + r) = f(x).

6. If r is odd, go back to step 1.
7. If a^r/2 ≡ -1 (mod N), go back to step 1.
8. The factors of N are gcd(ar/2 ± 1, N). We are done.

```julia
function blackBox(y, k, N)
      return mod(y^k, N)
end
    
```

```julia
function shor(t, N, y)
    length = BigInt(2)^BigInt(t) - 1
    sum=[]
    for k=0:length
        productOfStates = [k blackBox(y, k, N)]
        push!(sum, productOfStates)
    end
    normalizedFactor = 1/sqrt(2^t)
    return sum, normalizedFactor
end
```

```julia
function measurment(state)
    n = size(state,1)
    length = n
    eigenVec = rand(1:n)
     collapsedSecondRegister = state[eigenVec][2]
 #   collapsedSecondRegister = 4
    println(collapsedSecondRegister)
    postState = [];
    for i=1:length
        if (state[i][2] == collapsedSecondRegister)
            push!(postState, state[i][1])
        end    
    end
    reNormalizedFactor = 1/sqrt(size(postState,1))
    return postState, reNormalizedFactor
end
```

```julia
function IDFT(pState, reNorm)
    N_orig = 2^t
    N_prim = size(pState,1)
    constFactor = reNorm/sqrt(N_orig)
    sumOfStates = []
    idfState = []
    outState = []
    sum_l = 0.
    for j=1:N_prim
        for k=1:N_orig
            factor = exp(2*pi*1im*pState[j]*k/N_orig)*constFactor
            temp = [factor k]
            push!(sumOfStates, temp)
        end
        push!(idfState, sumOfStates)
        sumOfStates=[]
    end
    for m=1:N_orig
        for l=1:N_prim
            sum_l = sum_l + (idfState[l][m][1])
        end
        temp2 = [sum_l idfState[1][m][2]]
        push!(outState, temp2)
        sum_l = 0
    end
    return outState
end
```

```julia
idfStates=IDFT(postMeasurmentState, reNormalizedFactor);
```

```julia
function probDistro(idfStates)
    prob=[]
    for i=1:2^t
        push!(prob,abs2(idfStates[i][1]))
    end
    #println(prob);
    return prob
end 

```

```julia
function secondMeasurment(probArray)
    randomizedState=[]
    length = size(probArray,1)
    for i=1:length
        r = BigInt(rand(1:2^t))
        push!(randomizedState,r*probArray[i])
    end
    return findmax(randomizedState)[2]
end
```

```julia
prob = probDistro(idfStates);
```

```julia
finalValue = secondMeasurment(prob)
```

```julia
function contFraction(x,y)
    g = gcd(x,y)
    return x/BigInt(g), y/BigInt(g)
end
```

```julia
function findFactors(y, r)
    f = mod(y^(r/2)-1,N)
    d = gcd(BigInt(f),N)
    if d>=2
        u = d
        v = N/d
        return u,BigInt(v)
    else
        println("please run the algorithm again")
    end
end
```

```julia
function bin(x, l)
    x_bin = zeros(Int128,1,l)
    i = 0
    for i = 0:l-1
        x_bin[l-i] = mod(x,2)
        q = div(x,2)
        
        if q < 2 
            x_bin[l-i-1] = q
            return x_bin
        else
            i = i + 1
            x = q
        end
    end
    return x_bin
end
```
